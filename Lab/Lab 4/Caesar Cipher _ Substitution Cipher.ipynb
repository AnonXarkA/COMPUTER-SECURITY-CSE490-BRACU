```python
import pandas as pd
def frequency_table(text):
  text = text.lower()
  table = []
  count = 0
  for i in text:
    if i in 'qwertyuiopasdfghjklzxcvbnm':
      count += 1
  for i in range(97,123):
      table.append(text.count(chr(i))/count)
  index = [i[0] for i in sorted(enumerate(table), key=lambda x:x[1])][::-1]
  for i in index:
    print(chr(97+i)+":"+str(table[i]))
```

```python
ciphertext = "OF EKAHMGUKBHIA, B LWZLMOMWMOGF EOHITK OL B DTMIGR GY TFEKAHMOFU ZA CIOEI WFOML GY HSBOFMTVM BKT KTHSBETR COMI EOHITKMTVM, BEEGKROFU MG B YOVTR LALMTD; MIT WFOML DBA ZT LOFUST STMMTKL (MIT DGLM EGDDGF), HBOKL GY STMMTKL, MKOHSTML GY STMMTKL, DOVMWKTL GY MIT BZGXT, BFR LG YGKMI. MIT KTETOXTK RTEOHITKL MIT MTVM ZA HTKYGKDOFU MIT OFXTKLT LWZLMOMWMOGF."
frequency_table(ciphertext)
```
